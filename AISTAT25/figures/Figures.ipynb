```python
def cond1(x):
    if x>0.65625: return 1
    elif x>0.59375: return 0.1
    elif x>0.25: return 0.9
    else: return 0.1

def cond2(x):
    if x>0.6666: return 0.99
    else: return 0.01
    
def cond3(x):
    return x+sin(5*x)*0.1

def cond4(x):
    if x<0.26 and x>0.2: return 1
    else: return 0

def cond5(x):
    y=sin(10*x)+1
    if x>0.3 and x<0.6:
        y+=20
    return y/21-0.01
```

```python
def cond(x):
    return x+sin(5*x)*0.1

X=list(arange(0,1,0.001))
figure(figsize=[10,8])
plot(X, [2*cond(x)-1 for x in X],linewidth=3)
plot([0,1],[0.25,0.25],label="$+\gamma$")
plot([0,1],[-0.25,-0.25],label="$-\gamma$")
grid()
ylim(-1,1)
ylabel('$E[label| x]$',fontsize=16)
xlabel('x',fontsize=16)
title('Monotone Condition',fontsize=16)
legend(fontsize=16)
```

```python
def cond(x):
    y=sin(10*x)+1
    if x>0.3 and x<0.6:
        y+=20
    return y/21-0.01
    

X=list(arange(0,1,0.001))
figure(figsize=[10,8])
plot(X, [2*cond(x)-1 for x in X],linewidth=3)
plot([0,1],[0.25,0.25],label="$+\gamma$")
plot([0,1],[-0.25,-0.25],label="$-\gamma$")
grid()
ylim(-1,1)
ylabel('$E[label| x]$',fontsize=16)
xlabel('x',fontsize=16)
title('Massart Condition',fontsize=16)
legend(fontsize=16)
```
