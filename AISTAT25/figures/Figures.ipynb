```python
def cond(x):
    return x+sin(5*x)*0.1

X=list(arange(0,1,0.001))
figure(figsize=[10,8])
plot(X, [2*cond(x)-1 for x in X],linewidth=3)
plot([0,1],[0.25,0.25],label="$+\gamma$")
plot([0,1],[0,0],color='k',label="0")
plot([0,1],[-0.25,-0.25],label="$-\gamma$")
xticks([0.28, 0.41, 0.625], ['$\lambda_L$', '$\lambda$', '$\lambda_R$'],fontsize=16)

grid()
ylim(-1,1)
ylabel('$E[label| x]$',fontsize=20)
xlabel('x',fontsize=20)
title('(a)  Monotone Condition',fontsize=20)
legend(fontsize=16);
```

```python
def cond(x):
    y=sin(10*x)+1
    if x>0.3 and x<0.6:
        y+=20
    return y/21-0.01
    

X=list(arange(0,1,0.001))
figure(figsize=[10,8])
plot(X, [2*cond(x)-1 for x in X],linewidth=3)
plot([0,1],[0.25,0.25],label="$+\gamma$")
plot([0,1],[0,0],color='k',label="0")
plot([0,1],[-0.25,-0.25],label="$-\gamma$")
grid()
ylim(-1,1)
ylabel('$E[label| x]$',fontsize=20)
xlabel('x',fontsize=20)
title('(b)   Massart Condition',fontsize=20)
legend(fontsize=16)
```
